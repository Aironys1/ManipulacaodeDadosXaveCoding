### Criando um Data Frame
Podemos criar um Data Frame a partir de um dicionário, onde cada chave possui uma lista de elementos de igual tamanho <br>
As chaves representam as colunas e cada um dos valores de sua lista representa o valor da linha correspondente para aquela coluna.


### Pefil do Analista

```python
funcionarios = pd.DataFrame({
    'nome': ['Aironys Garrido', 'Samantha Silva', 'Vanessa Castelo'],
    'Cargo': ['Analysta Junior', 'Supervosora Tec', 'Departamento Pessoal'],
    'NumeroDaSorte':[7,5,8],
    'dataNascimento': ['05/12/1990', '22/04/1996','15/10/1985'],
    'salario': [2.817, 5.250, 3.200],
    'trabalhando': [True, True, False]

})

display(funcionarios)
```

```python
funcionarios.info()
```

### Renomenando as colunas de um Data Frame
O atributo DataFrame.columns retorna uma (lista) com nomes de todas as colunas do data Frame 

```python
funcionarios.columns
```

```python
type(funcionarios)
```

```python
# list(funcionarios)
```

```python
display(funcionarios)
```

### Agora vamos renomear as colunas

```python
funcionarios.rename(columns={
    'nome':'Nome Completo',
    'Cargo':'Função',
    'dataNascimento':'Nascimento',
    'salario':'Salário',
    'trabalhando':'Atividade'
}, inplace=True)

display(funcionarios)
```

```python
# Renomendo só as colunas 
funcionarios.columns
```

```python
funcionarios.columns = ['NOME', 'CARGO', 'NÚMERO DA SORTE', 'NASCIMENTO', 'SALÁRIO', 'ATIVIDADE']
```
